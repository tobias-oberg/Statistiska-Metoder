```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
Auto = pd.read_csv("Auto.csv")
Auto
```

```python
# a) Which of the predictors are quantitative(numerical) and which are qualitative(categorial)? 

print(Auto.dtypes)

quantitative = []
qualitative = []

for col in Auto.columns:
    if Auto[col].dtype in ["int64", "float64"]:
        quantitative.append(col)
    else:
        qualitative.append(col)

print("Quantitative predictors:", quantitative)
print("Qualitative predictiors:", qualitative)

```

```python
# b) What is the range of each quantitative predictor? You can answer this using the min() and max() methods in numpy.

print("The range of each quantitative predictor:")
for col in quantitative:
    min_value = np.min(Auto[col])
    max_value = np.max(Auto[col])
    print(f"{col}: Min: {min_value}, Max: {max_value}, Range: {max_value - min_value}")

```

```python
# c) What is the mean and standard deviation of each quantitative predictor?

print("The mean and standard deviation of each quantitative predictor:")
for col in quantitative:
    mean_value = np.mean(Auto[col])
    standard_deviation = np.std(Auto[col])
    print(f"{col}: Mean value: {mean_value:.2f}, Standard deviation: {standard_deviation:.2f}")
```

```python
# d) Now remove the 10th through 85th observations. 
# What is the range, mean, and standard deviation of each predictor in the subset of the data that remains?

Auto_filtered = Auto.drop(index=range(9,85))
for col in quantitative:
    min_value = np.min(Auto_filtered[col])
    max_value = np.max(Auto_filtered[col])
    mean_value = np.mean(Auto_filtered[col])
    standard_deviation = np.std(Auto_filtered[col])
    print(f"{col}: Range: {max_value - min_value}, Mean: {mean_value:.2f}, Standard deviation: {standard_deviation:.2f}")


```

```python
# e) Using the full data set, investigate the predictors graphically,using scatterplots or other tools of your choice. 
# Create some plots highlighting the relationships among the predictors. 
# Comment on your findings.
plt.figure(figsize=(8,5))
plt.scatter(Auto["weight"], Auto["mpg"], alpha=0.5,)
plt.title("Weight vs mpg")
plt.xlabel("Weight")
plt.ylabel("miles per gallon")
plt.grid(True)
plt.show()
```
